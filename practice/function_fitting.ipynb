```python
from kan import *
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
```

```python
model = KAN(width=[2, 4, 2, 1], grid=5, k=3, seed=1, device=device)
f = lambda x: torch.exp(torch.sin(torch.cos(x[:,[0]]) + x[:,[1]]**2))
dataset = create_dataset(f, n_var=2, device=device)
```

```python
model.plot()
```

```python
model.fit(dataset, opt="LBFGS", steps=500, update_grid=True, start_grid_update_step=10, stop_grid_update_step=150, grid_update_num=10, log=25);
```
